#### Creating 2 new columns

```python
epl["Mins Per Match"]=(epl["Mins"]/epl["Matches"]).astype(int)
epl["Goals Per Match"]=(epl["Goals"]/epl["Matches"]).astype(float)
epl.head()
```

***
## Data Analysis

### Total Goals

```python
Total_Goals=epl.Goals.sum()
print("Total Goals in the EPL 2020-2021 =", Total_Goals)
```

***
### Total Penalty Goals

```python
Total_Penalty_Goals=epl["Penalty_Goals"].sum()
print("Total Penalty Goals in the EPL 2020-2021 =",Total_Penalty_Goals)
```

***
###  Total Penalty Attempts

```python
 Total_Penalty_attempts=epl["Penalty_Attempted"].sum() #Total_Penalty_attempts= Penalty_missed + Total_Penalty_Goals
Penalty_missed=Total_Penalty_attempts-Total_Penalty_Goals
print("Total Penalty Attempts =",Total_Penalty_attempts)
print("Penalty Attempts missed to goal=",Penalty_missed)
```

***
### Penalty missed vs Penalty scored

```python
plt.figure(figsize=(13,6))
data=[Penalty_missed,Total_Penalty_Goals]
labels=["Penalties Missed","Penalties Scored"]
color=sns.color_palette("hsv")
plt.pie(data,labels=labels,colors=color,autopct="%.0f%%")
plt.show()
```

#### Observation from Penalty Goals:
- Most of the Penalties were scored with the percentage of 82%
- Around only 18% of the total penalties were missed

### Players based on the age group

```python
Under20=epl[epl["Age"]<=20]
age20_25=epl[(epl["Age"]>20) & (epl["Age"]<=25)]
age25_30=epl[(epl["Age"]>25) & (epl["Age"]<=30)]
age30=epl[epl["Age"]>30]
```

```python
data=np.array([Under20["Name"].count(),age20_25["Name"].count(),age25_30["Name"].count(),age30["Name"].count()])
plt.figure(figsize=(10,8))
mylabel=["<=20",">20 & <=25",">25 & <=30",">30"]
color=sns.color_palette("prism")
plt.title("Total Players with Age",fontsize=20)
plt.pie(data,labels=mylabel,autopct="%.0f%%")
plt.show()
```

#### Observation
In overall EPL, 
- Players between the age group of 25 and 30 are high with 37% of the total players.
- Players between the age group of 20 adn 25 are the second highest with the 35% of the total players
- Players under 20 age group are least with the 15% of the total players

### Goals with assist and Goals without assist

```python
Total_assist=epl.Assists.sum()
data=[Total_Goals-Total_assist,Total_assist]
label=["Goals without Assist","Goals with Assist"]
plt.figure(figsize=(12,6))
plt.pie(x=data,labels=label,colors=sns.color_palette("Set1"),autopct="%.0f%%");
plt.show()
```

#### Observation
- Most of the goals are with Assist (69% of total)
- 31% of the goals are without Assist
